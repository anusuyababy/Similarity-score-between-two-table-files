```python
from docx2csv import extract_tables, extract
tables = extract_tables('/content/lists.docx')
```

```python
import pandas as pd
from docx import Document
document = Document("/content/lists.docx")
```

```python
def read_docx_table(document,table_num):
  table = document.tables[table_num-1]
  data = [[cell.text for cell in row.cells] for row in table.rows]
  df = pd.DataFrame(data)
  return df
```

```python
table_num=1
df = read_docx_table(document,table_num)
df.head()
```

```python
q1 = df.copy()
```

```python
table_num=2
df1 = read_docx_table(document,table_num)
df1.head()
```

```python
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
sen_embeddings = np.asarray(model.encode(df[1]))
sen_embeddings.shape
```

```python
cou_embeddings = np.asarray(model.encode(df1[1]))
cou_embeddings.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#let's calculate cosine similarity for sentence 0:
similarities =cosine_similarity(
    cou_embeddings, sen_embeddings)
```

```python
similarities
```

```python
S = pd.DataFrame(similarities, columns=['P1','P2', 'P3', 'P4', 'P5', 'P6'])
S.head()
```

```python
S.tail()
```

```python
df2 = pd.concat([df1[0], S], axis=1)
df2.head()
```

```python
df2.set_index(0, inplace=True)
df2.head()
```
